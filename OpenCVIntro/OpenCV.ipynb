### Affine Transformations

```python
img = cv2.imread('Data/glass.jpg')
rows,cols,ch = img.shape

pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])

M = cv2.getAffineTransform(pts1,pts2)

dst = cv2.warpAffine(img,M,(cols,rows))
plt.imshow(dst)
```

```python
img = cv2.imread('Data/glass.jpg')
pts1 = np.float32([[75,20],[145,20],[85,200],[140,200]])
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])

M = cv2.getPerspectiveTransform(pts1,pts2)

dst = cv2.warpPerspective(img,M,(300,300))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
```

```python
img = cv2.imread('Data/glass.jpg')
M = np.float32([[1,0,80],[0,1,70]])
dst = cv2.warpAffine(img,M,(cols,rows))
plt.imshow(dst)
```

```python
img = cv2.imread('Data/glass.jpg')
rows,cols,_ = img.shape

M = cv2.getRotationMatrix2D((cols/2,rows/2),-45,1)
dst = cv2.warpAffine(img,M,(cols,rows))
plt.imshow(dst)
```

```python
img = cv2.imread('Data/j.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
kernels = np.ones((3,3), np.uint8)
erosion = cv2.erode(img, kernels, iterations=2)
plt.imshow(erosion)
```

```python
img = cv2.imread('Data/lenna.jpeg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
kernels = np.ones((5,5), np.uint8)
erosion = cv2.dilate(img, kernels, iterations=1)
plt.imshow(erosion)
```
